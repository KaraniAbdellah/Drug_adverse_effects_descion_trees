## Get The Data From API

```python
# Import Packages
import requests
import json
from openpyxl import Workbook
import pandas as pd
from pprint import pprint
```

```python
# API-ENDPOINT
numberOfDocument = 1
numberOfRow = 1000
URL = f"https://api.fda.gov/drug/event.json"
```

```python
# Load Data
req = requests.get(url=URL, params={'limit': numberOfDocument})
data = req.json()
dict_data = data["results"][0]

df1 = pd.json_normalize(data=dict_data) # normalize json data
df2 = df1.melt() # flip the data vertically
df2
```

```python
# Export it to CSV
df1.to_csv("data.csv", index=False) # remove index (0, 1, 2, ... n)
```
